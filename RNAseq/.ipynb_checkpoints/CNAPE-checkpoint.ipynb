```python
import bioNet
class myClassifier(xgb.XGBClassifier):
    def __init__(self, subset=None, objective="binary:logistic", **kwargs):
        super().__init__(objective=objective, **kwargs)
        self.params_bk = kwargs
        self.params_bk['objective']=objective
        try:
            self.subset = sorted(list(set(subset)))
        except:
            self.subset = None

    def fit(self, X, y, sample_weight=None, eval_set=None, eval_metric=None, early_stopping_rounds=None, verbose=True, xgb_model=None, sample_weight_eval_set=None, callbacks=None):
        self.newSubset = []
        if self.subset == 0:
            print('raw')
        else:
            for i in self.subset:
                if i in X.columns:
                    self.newSubset.append(i)
            X = X[self.newSubset]
            print(len(X.columns))
        super().fit(X, y, sample_weight, eval_set, eval_metric, early_stopping_rounds, 
                    verbose, xgb_model, sample_weight_eval_set, callbacks)
    
    def predict(self, data, output_margin=False, ntree_limit=None, validate_features=True, base_margin=None):
        if self.subset == 0:
            print('')
        else:
            data = data[self.newSubset]
            print(len(data.columns))
        return super().predict(data, output_margin, ntree_limit, validate_features, base_margin=None)
    
    def predict_proba(self, data, ntree_limit=None, validate_features=True):
        if self.subset == 0:
            print('')
        else:
            data = data[self.newSubset]
            print(len(data.columns))
        return super().predict_proba(data, ntree_limit, validate_features)
    
    def get_xgb_params(self):
        return myc.params_bk
    def get_params(self):
        return myc.params_bk
```

```python
df = pd.DataFrame()
for i in tcgaCNA.columns:
    if i[:3]!='chr':
        print(i)
        h1=bioNet.GRCh37()
        h1.load('./dbs/Homo_sapiens.GRCh37.75.genes.typeNlengthNCoord.txt')
        a1 = h1.getNeighbor(i)

        h2=bioNet.string()
        h2.load('./dbs/9606.protein.info.v11.0.txt',\
            './dbs/9606.protein.links.v11.0.txt')
        a2 = h2.getNeighbor(i,0)

        h3=bioNet.kegg()
        h3.load('./dbs/c2.cp.kegg.v7.0.symbols.gmt')
        a3 = h3.getNeighbor(i)

        y = 1*(tcgaCNA[i]!='no')
        myc = myClassifier(objective='binary:logistic', booster='gbtree', max_depth=1, scale_pos_weight=len(y)/sum(y),
                          early_stopping_rounds=10, random_state=2020, subset=set(list(a1)+list(a2)+list(a3)), missing=None)
        myc.fit(X=tcgaExp,y=y)
        prob = []
        for j in myc.predict_proba(bjniExp):
            prob.append(j[1])
        df[i] = prob
df.index = bjniExp.index
```
